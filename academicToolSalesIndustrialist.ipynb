```python
import pandas as pd
```

```python
book_sales_df = pd.read_excel('/content/Book Sales .xlsx')
```

```python
book_sales_df
```

```python
book_sales_df.columns
```

```python
#copy to a different df to edit
book_sales_df_copy = book_sales_df.copy()
```

# Replace the headers with the first column

```python
# Assign the 1st row as the header.
book_sales_df_copy.columns = book_sales_df_copy.iloc[0]

# Remove the 2nd row.
book_sales_df_copy = book_sales_df_copy[1:]

# Reset the index.
book_sales_df_copy = book_sales_df_copy.reset_index(drop=True)
```

```python
book_sales_df_copy.columns
```

```python
book_sales_df_copy
```

```python
book_sales_df_copy.info()
```

```python
book_sales_df_copy['price'] = pd.to_numeric(book_sales_df_copy['price'])
```

```python
# Convert 'Revenue' and 'Unit sold' columns to numeric type
book_sales_df_copy[['Revenue', 'Unit sold']] = book_sales_df_copy[['Revenue', 'Unit sold']].apply(pd.to_numeric, errors='coerce')
```

```python
!pip install matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
total_revenue = book_sales_df_copy.Revenue.sum()
total_revenue
```

```python
total_sold_unit_overtime = book_sales_df_copy["Unit sold"].sum()
total_sold_unit_overtime
```

```python
book_sales_df_copy['date'] = pd.to_datetime(book_sales_df_copy['date'])
```

```python
import calendar
month_order = list(calendar.month_abbr[1:])
sales_for_each_month = book_sales_df_copy.groupby('month')['Revenue'].sum()
```

```python
# Plot the monthly revenue using a bar chart
plt.figure(figsize=(10, 6))
sales_for_each_month = sales_for_each_month.reindex(month_order)
plt.bar(sales_for_each_month.index, sales_for_each_month.values)

plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.title("Monthly Revenue Trend")
plt.xticks(rotation=45)
plt.tight_layout()
# Disable scientific notation on y-axis
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```

```python
# Sales over time
sales_over_time = book_sales_df_copy.groupby('date')['Revenue'].sum()
sales_over_time
```

```python
sales_over_time.sort_values(ascending=False)
```

```python
!pip install plotly
import plotly.express as px

fig = px.line(x=sales_over_time.index, y=sales_over_time.values, title='Sales Over Time')
fig.update_layout(xaxis_title="Date", yaxis_title="Total Revenue")
fig.show()
```

# Sales Distribution by Product

Grouped each products as a series and summed Revenue gotten from each product.

```python
# Sales distribution by product category
sales_by_category = book_sales_df_copy.groupby('Products')['Revenue'].sum()

fig, axes = plt.subplots(1, 2, figsize=(16, 5)) # Increased figure size

axes[0].set_title('Barplot sales dist.', fontsize=16) # Increased title font size
sns.barplot(x=sales_by_category.index, y=sales_by_category.values, ax=axes[0])
axes[0].set_xlabel('Product', fontsize=16)
axes[0].set_ylabel('Revenue generated', fontsize=16)
axes[0].tick_params(axis='both', which='major', labelsize=14)
axes[0].ticklabel_format(style='plain', axis='y')


axes[1].set_title('Piechart sales dist.', fontsize=16)
axes[1].pie(sales_by_category.values, labels=sales_by_category.index, autopct='%1.1f%%')
axes[1].tick_params(axis='both', which='major', labelsize=14) # Increased tick label font size

plt.tight_layout()
```

The graph shows that Text books was by a large mile the most sold academic tool, followed by the calculator and dictionary which were not separated by much margin. Not much people bought the exercise books which generated the least revenue, followed by Mathset when compared to the rest of academic tool in sales distribution.
